```python
import json

# Search term
search_term = input("Enter the search term: ")

# Search for matches based on paramName
matches = [info for info in climateDT_params if search_term.lower() in info["paramName"].lower()]

if len(matches) == 0:
    print("No matches found.")
else:
    # Print paramName and param of each match
    print("Matching Parameters:")
    for index, match in enumerate(matches):
        print(f"{index + 1}. paramName: {match['paramName']}, param: {match['param']}")

    # Ask user to select a match
    selected_index = input("Enter the index of the parameter you want to select: ")
    try:
        selected_index = int(selected_index)
        if 1 <= selected_index <= len(matches):
            selected_entry = matches[selected_index - 1]
            # Assuming `selected_entry` is the dictionary you want to pretty print
            selected_entry_json = json.dumps(selected_entry, indent=4)
            print(selected_entry_json)
        else:
            print("Invalid index entered.")
    except ValueError:
        print("Invalid input. Please enter a number.")
```
